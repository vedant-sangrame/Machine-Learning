```python
X = df.drop(columns=['target_class'])
y = df['target_class']
```

```python
new_column_names = {
    'Mean of the integrated profile': 'Mean_Int_Profile',
    'Standard deviation of the integrated profile': 'StdDev_Int_Profile',
    'Excess kurtosis of the integrated profile': 'ExKurt_Int_Profile',
    'Skewness of the integrated profile': 'Skew_Int_Profile',
    'Mean of the DM-SNR curve': 'Mean_DM_SNR',
    'Standard deviation of the DM-SNR curve': 'StdDev_DM_SNR',
    'Excess kurtosis of the DM-SNR curve': 'ExKurt_DM_SNR',
    'Skewness of the DM-SNR curve': 'Skew_DM_SNR',
    'target_class': 'Target'
}

df = df.rename(columns=new_column_names)
display(df.head())
```

```python
X = pd.get_dummies(X,dtype = int)
```

```python
X = df.iloc[:,3:-1]
X = pd.get_dummies(X, dtype=int)
X
```

```python
y=df.iloc[:,-1]
print(y)
```

```python
X.shape
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=42)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
svc = SVC()
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
```

```python
svc = SVC(C=100.0)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
svc = SVC(C=1000.0)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
linear_svc = SVC(kernel='linear', C=1.0)
linear_svc.fit(X_train,y_train)
y_pred = linear_svc.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
linear_svc = SVC(kernel='linear', C=100.0)
linear_svc.fit(X_train,y_train)
y_pred = linear_svc.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```
