```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

path = "/content/drive/MyDrive/ML_dataset/Spotify_Songs_Dataset - Spotify_Songs_Dataset.csv"
df = pd.read_csv(path)

df = df[['popularity', 'danceability']].dropna()

df = df.sample(n=500, random_state=42)   # IMPORTANT

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

plt.figure(figsize=(8, 5))
dendrogram(linkage(X_scaled, method='ward'))

plt.title("Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
cluster = hc.fit_predict(X_scaled)
```

```python
df['Cluster'] = cluster
print(df.head())
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(x='popularity',
                y='danceability',
                hue='Cluster',
                data=df,
                palette='Set1'
)

plt.title("Hierarchical Clustering of Spotify Songs")
plt.show()
```
